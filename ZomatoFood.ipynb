```python
def read_data_from_csv():
    hotels = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HiCounselor/Zomato/zomato.csv')
    return hotels
```

```python
hotels = read_data_from_csv()
df = hotels.copy()
hotels.head()
```

```python
hotels.columns = hotels.columns.str.replace(' ', '_')
hotels.drop(['phone','address'], inplace=True,axis=1)
```

```python
    # task2: rename columns,  only these columns are allowed in the dataset
    # 1.	Id not available
    # 2.	Name
    # 3.	online_order
    # 4.	book_table
    # 5.	rating
    # 6.	votes
    # 7.	location
    # 8.	rest_type
    # 9.	dish_liked
    # 10.	cuisines
    # 11.	approx_cost
    # 12.	type

hotels.rename(columns={'name':'Name','rate': 'rating','approx_cost(for_two_people)': 'approx_cost','listed_in(type)':'type'},inplace=True)
print(hotels.info())
print(hotels.head())
```

```python
# task3: handle  null values of each column
hotels.isna().sum()
```

```python
hotels.describe(include='all').T
```

```python
hotels.dropna(subset=['Name'],axis=0,inplace=True)
hotels['online_order'].fillna('NA',inplace=True)
hotels['book_table'].fillna('NA',inplace=True)
hotels['rating'].fillna(0,inplace=True)
hotels['votes'].fillna(0,inplace=True)
hotels['location'].fillna('NA',inplace=True)
hotels['rest_type'].fillna('NA',inplace=True)
hotels['dish_liked'].fillna('NA',inplace=True)
hotels['cuisines'].fillna('NA',inplace=True)
hotels['approx_cost'].fillna(0,inplace=True)
hotels['type'].fillna('NA',inplace=True)
hotels.isna().sum()
```

```python
# task4 #find duplicates in the dataset
hotels.duplicated().sum()
hotels.drop_duplicates(inplace=True)
hotels.reset_index(drop=True,inplace=True)
```

```python
hotels.info()
```

```python
# task5 removing irrelevant text from all the columns
for column in hotels.columns:
        hotels = hotels[hotels[column].str.contains('RATED|Rated') == False]
```

```python
# task6: check for unique values in each column and handle the irrelevant values
hotels['online_order'].unique()
index = hotels[~hotels['online_order'].isin(['No', 'Yes'])].index
hotels.drop(index, inplace = True)
hotels['rating'].replace('NEW',0,inplace=True)
hotels['rating'].replace('-',0,inplace=True)
hotels['rating'].replace(to_replace='/5',value='',regex=True,inplace=True)
```

```python
# task7: remove the unknown character from the dataset and export it to "zomatocleaned.csv"
# remove unknown character from dataset
hotels['name'] = hotels['name'].str.replace('[Ãx][^A-Za-z]+','',regex=True)

```

```python
# export cleaned Dataset to newcsv file named "zomatocleaned.csv"
hotels.to_csv('zomatocleaned.csv')
```
